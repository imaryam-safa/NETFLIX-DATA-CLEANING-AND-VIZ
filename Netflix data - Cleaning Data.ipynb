```python
#Importing Libraries
import pandas as pd
import os
```

```python
#Path of the original file and the new file
path = "C:/Users/Safa/Desktop/data analytics"
data_dir = os.path.join(path, "netflix_titles.xlsx")
data_output= os.path.join(path,"netflix_titles_new.xlsx")
```

```python
#Checking if the file has multiple sheets
df = pd.ExcelFile(data_dir)
df.sheet_names
```

```python
#Reading the file 
df = pd.read_excel(data_dir, date_parser=True)
```

```python
#Displaying the first 5 columns
df.head()
```

```python
#Checking the nulls
df.info()
```

```python
#Checking for the total number of null values
df.isnull().sum()
```

```python
#Replacing null values with placeholders, or deleting the rows altogether
df.director.fillna("Not Given",inplace=True)
df.cast.fillna("Not Given",inplace=True)
df.country.fillna("Not Given",inplace=True)
df.dropna(axis=0, subset=['date_added', 'rating', 'duration'], inplace=True)
```

```python
#checking for null values after cleaning
df.isnull().sum()
```

```python
#Deleting unneeded columns which many not help with the viz
df.drop(['cast', 'description'],axis=1, inplace=True)
```

```python
#checking if the columns have been deleted
df.info()
```

```python
#Checking the country column
df.country.unique()
```

```python
#since there are multiple countries in each row, we will delete values after the first comma
df['country'] = df['country'].str.split(',').str[0]
```

```python
#Checking if multiple values have been deleted
df.country.unique()
```

```python
#Checking the first 5 rows again
df.head()
```

```python
#Exporting file to excel 
df.to_excel(data_output,columns=['show_id','type','title','director','country','date_added','release_year','rating','duration','listed_in'])
```
